```python
import scipy.stats as st
from scipy import stats
```

```python
stat, p = st.ttest_1samp(a = salary_high['average_montly_hours'], popmean = salary_low['average_montly_hours'].mean())
print(f'Stat ={stat:.3f}, p = {p:.3f}')
if p > 0.05:
  print('Сотрудники с высоким окладом , вероятно,проводят больше времени на работе, чем сотрудники с низким окладом')
else:
  print('сотрудники с высоким окладом проводят на работе меньше времени, чем сотрудники с низким окладом')
```

```python
resigned = resigned[[ 'satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company'\
                    , 'Work_accident', 'promotion_last_5years']]
```

```python
resigned
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(resigned, resigned['number_project'], random_state=42)
```

```python
y_train
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(x_train, y_train)
```

```python
lda.predict(x_test)
```

```python
result = pd.DataFrame([y_test, lda.predict(x_test)]).T
result
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, lda.predict(x_test))
```
